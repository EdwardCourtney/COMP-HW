```python
#W4A8
def reverse(n: int):
    n = str(n)
    A = list(n)
    ans = ""
    for i in range(len(A)-1, -1, -1):
        ans += A[i]
    ans = int(ans)
    return ans
def pali(n: int) -> bool:
    n = str(n)
    A = list(n)
    ans = True
    for i in range(len(A)//2):
        if A[i] != A[len(A)-1-i]:
            ans = False
            break
    return ans

n = int(input())
ans = 0
while (not pali(n)):
    n += reverse(n)
    ans += 1
print(ans, " ",n)
    
```

```python
#W4A9
def distinct(n: int):
    n = str(n)
    A = list(n)
    for i in range(len(A)):
        if n.count(A[i]) > 1:
            return False
    return True
        
n = int(input())
i = 1
while i*i<=n:
    if distinct(i*i):
        print(i*i, end=" ")
    i += 1
```

```python
#W4A10
def collazt(n):
    ans = 1
    while n != 1:
        if n%2 == 0:
            n = n/2
        else:
            n = 3*n + 1
        ans += 1
    return ans

n = int(input())
l = -1
x = -1
for i in range(1, n+1):
    c = collazt(i)
    if c > l:
        l = c
        x = i
print(x, l)    
```
