```python
#W7A1
def bs(A, n):
    l = 0
    r = len(A)-1
    while l <= r:
        m = (l+r)//2
        if A[m] == n:
            return m+1
        elif A[m] < n:
            l = m+1
        else:
            r = m-1
    return -1

A = list(map(int, input().split()))
n = int(input())
print(bs(A, n))
```

```python
#W7A2
A = list(map(int, input().split()))
x = int(input())
D = {}
for i in range(len(A)):
    if A[i] not in D:
        D[A[i]] = 0
    D[A[i]] += 1

if x not in D:
    print(0)
else:
    print(D[x])
```

```python
#W7A3
def bubbleSort(A: list):
    for i in range(len(A)):
        unsorted = True
        for j in range(1, len(A)-i):
            if (A[j] < A[j-1]):
                unsorted = False
                t = A[j]
                A[j] = A[j-1]
                A[j-1] = t
        if unsorted:
            return A
    return A
   
A = list(map(int, input().split()))
print(*bubbleSort(A))
```

```python
#W7A4
A = list(map(int, input().split()))
D = {}
for i in range(len(A)):
    if A[i] not in D:
        D[A[i]] = 0
    D[A[i]] += 1

a, b = (0,-10000000)
for x, y in D.items():
    if y > b:
        b = y
        a = x

print(f"{a} xuat hien nhieu nhat, som nhat, {b} lan")
```

```python
#W7A5
A = list(map(int, input().split()))
X = int(input())
D = {}
ans = 0
for i in range(len(A)):
    if A[i] not in D:
        D[A[i]] = 0
    D[A[i]] += 1

for e in D:
    if X - e in D:
        ans += D[e]*D[X-e]
print(ans)    
```
