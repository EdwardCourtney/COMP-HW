```python
#W7A1
def bs(A, n):
    l = 0
    r = len(A)-1
    while l <= r:
        m = (l+r)//2
        if A[m] == n:
            return m+1
        elif A[m] < n:
            l = m+1
        else:
            r = m-1
    return -1

A = list(map(int, input().split()))
n = int(input())
print(bs(A, n))
```
