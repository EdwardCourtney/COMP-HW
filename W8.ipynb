```python
#W8A1
class Cylinder:
    Pi = 3.14
    def __init__(self, r, h):
        self.radius = r
        self.height = h
    def Volume(self):
        return self.Pi * ((self.radius)**2) * self.height
    def SurfaceArea(self):
        return self.Pi * 2 * self.radius * self.height + self.radius * 2 * self.Pi * self.radius

a, b = map(int, input().split())
C = Cylinder(a, b)
print(f"{C.SurfaceArea():.2f} {C.Volume():.2f}")
```

```python
#W8A3
class Point():
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def centre(self):
        if self.x == 0 and self.y == 0:
            print("Is centre")
        else:
            print("Is not centre")
    def yAxis(self):
        if self.x == 0:
            print("Is on Y-Axis")
        else:
            print("Is not on Y-Axis")
    def xAxis(self):
        if self.y == 0:
            print("Is on X-Axis")
        else:
            print("Is not on X-Axis")
    def distance(self, other):
        ans = ((self.x - other.x)**2 + (self.y - other.y)**2)**0.5
        print(f"Distance = {ans}")
A = Point(1, 1)
B = Point(0, 1)
C = Point(0, 0)
A.centre()
B.yAxis()
Point.centre(C)
A.distance(B)
Point.distance(A, B)
```

```python
#W8A5
class ShoppingCart():
    def __init__(self):
        self.L = list()
    def add(self, name, price):
        self.L.append({name: price})
    def remove(self, name):
        for i in range(len(self.L)):
            if name in self.L[i]:
                self.L.pop(i)
                break
    def empty(self):
        if len(self.L) == 0:
            return True
        else:
            return False
    def total(self):
        ans = 0
        for i in range(len(self.L)):
            for x in self.L[i].values():
                ans += x
        return ans
    def show(self):
        return self.L
    def clear(self):
        self.L = []
print("Welcome to Shopping Cart")
S = ShoppingCart()
while (True):
    print('''Command Guide:
        - 2. Add item
        - 3. Remove item
        - 4. Check if shopping cart is empty
        - 5. Calculate total price
        - 6. Show all items and prices
        - 7. Clear shopping cart''')
    print("Please input command")
    I = int(input())
    if I == 2:
        print("Please input name and price in one line")
        N, P = input().split()
        P = int(P)
        S.add(N, P)
    if I == 3:
        print("Please input item name for removal")
        N = input()
        S.remove(N)
    if I == 4:
        if S.empty():
            print("Your shopping cart is empty")
        else:
            print("Your shopping cart is not empty")
    if I == 5:
        print(f"Total price: {S.total()}")
    if I == 6:
        print("Your items and prices:")
        for i in range(len(S.L)):
            print(f"{i+1}. {list(S.L[i].keys())[0]} - {list(S.L[i].values())[0]}")
    if I == 7:
        S.clear()
        print("Your shopping cart has been cleared")
    print("Do you want to exit program? Input Yes to exit or Enter to continue")
    A = input()
    if (A.lower() == "yes"):
        print("Thank you for using Shopping Cart")
        break
    else:
        print("----------------")
        
```

```python
#W8A7
class Fraction():
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def __gcd(self, a, b):
        for i in range(1, a*b+1):
            if i % a == 0 and i % b == 0:
                return i
            
    def __lmc(self, a, b):
        return (a*b)//self.__gcd(a, b)
            
    def __denominator(self, other):
        k = self.__gcd(self.b, other.b)
        (self.a, self.b, other.a, other.b) = (self.a*(k//self.b), k, other.a*(k//other.b), k)
        
    def __shorten(self, a, b):
        k = self.__lmc(a,b)
        return (a//k, b//k)
        
    def __add__(self, other):
        self.__denominator(other)
        return self.__shorten(self.a + other.a, self.b)
    
    def __sub__(self, other):
        self.__denominator(other)
        return self.__shorten(self.a - other.a, self.b)
    
    def __mul__(self, other):
        return self.__shorten(self.a * other.a, self.b * other.b)
    
    def __truediv__(self, other):
        return self.__shorten(self.a * other.b, self.b * other.a)


a1, b1, e, a2, b2 = input().split()
a1, a2, b1, b2 = map(int, [a1, a2, b1, b2])

A, B = (0, 0)
F1 = Fraction(a1, b1)
F2 = Fraction(a2, b2)

if e == "+":
    A, B = F1 + F2
elif e == "-":
    A, B = F1 - F2
elif e == "*":
    A, B = F1 * F2
else:
    A, B = F1 / F2

print(f"{A}/{B}")
```

```python
#W8A8
class BankAccount():
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, x):
        self.balance += x
    def withdraw(self, x):
        if not (x > self.balance):
            self.balance -= x

A, t = input().split()
t = float(t)
n = int(input())
Ba = BankAccount(A, t)

for i in range(n):
    S = input()
    c, x = S.split()
    x = float(x)
    if c == "DEPOSIT":
        Ba.deposit(x)
    else:
        Ba.withdraw(x)
print(f"{Ba.owner}'s balance is {Ba.balance:.2f}")
```

```python
#W8A9
class Student():
    def __init__(self, name):
        self.name = name
        self.GPA = {}
    
    def add(self, subject, score):
        self.GPA[subject] = score
    
    def avg(self):
        ans = 0
        i = 0
        for x in self.GPA:
            ans += self.GPA[x]
            i += 1
        return ans/i
    
    def rank(self):
        A = self.avg()
        if A >= 8:
            return "Excellent"
        elif A < 8 and A >= 6:
            return "Average"
        else:
            return "Poor"

S = Student(input())
n = int(input())
for i in range(n):
    Name, Score = input().split()
    Score = float(Score)
    S.add(Name, Score)
print(f"{S.name} {S.avg():.2} {S.rank()}")

    
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self. author = author
        self.year = year
    
class Library():
    def __init__(self):
        self.Lib = []
        
    def add(self, book):
        self.Lib.append(book)
        
    def count_by_name(self, name):
        ans = 0
        for i in range(len(self.Lib)):
            self.Lib[i].author == name
            ans += 1
        return ans
    
    def find_by_year(self, year):
        ans = 0
        for i in range(len(self.Lib)):
            self.Lib[i].author == year
            ans += 1
        return ans

n = int(input())
L = Library()
for i in range(n):
    command, temp = input().split()
    if command == "ADD":
        title, author, year = temp.split(";")
        B = Book(title, author, year)
        L.add(B)
    elif command == "COUNT":
        print(L.count_by_name(temp))
    else:
        print(L.find_by_year(temp))
```

```python
#W8A11
from math import acos, pi
class Point():
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

P = []
for i in range(4):
    x, y, z = map(float, input().split())
    P.append(Point(x, y, z))
    
V1 = Point(P[0].x-P[1].x, P[0].y-P[1].y, P[0].z-P[1].z)
V2 = Point(P[2].x-P[1].x, P[2].y-P[1].y, P[2].z-P[1].z)
V3 = Point(P[3].x-P[2].x, P[3].y-P[2].y, P[3].z-P[2].z)

P1 = Point(V1.y*V2.z - V2.y*V1.z, V1.x*V2.z - V2.x*V1.z, V1.x*V2.y - V2.x*V1.y)
P2 = Point(V3.y*V2.z - V2.y*V3.z, V3.x*V2.z - V2.x*V3.z, V3.x*V2.y - V2.x*V3.y)

ANS = acos(abs(P1.x*P2.x + P1.y*P2.y + P1.z*P2.z)/((((P1.x)**2 + (P1.y)**2 + (P1.z)**2)**0.5) * ((P2.x)**2 + (P2.y)**2 + (P2.z)**2)**0.5))
ANS = (ANS*180)/pi
if ANS > 90:
    ANS -= 90
print(f"{ANS:.2f}")
```

```python
#W8A12
class Complex():
    def __init__(self, re, im):
        self.re = re
        self.im = im
    def product(self, other):
        return (self.re*other.re - other.im*self.im, self.re*other.im + self.im*other.re)

a, b = map(int, input().split())
c, d = map(int, input().split())

A = Complex(a, b)
B = Complex(c, d)

t1, t2 = A.product(B)
print(f"{t1} {t2}")
```

```python
#W8A13
class Triangle():
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3
    def getArea(self):
        if not (self.e1 + self.e2 > self.e3 and self.e2 + self.e3 > self.e1 and self.e1 + self.e3 > self.e2):
            return -1
        else:
            p = (self.e1 + self.e2 + self.e3)/2
            return (p*(p-self.e1)*(p-self.e2)*(p-self.e3))**0.5

a, b, c = map(int, input().split())
T = Triangle(a, b, c)
if T.getArea() == -1:
    print("Invalid")
else:
    print(f"{T.getArea():.2f}")
```
