```python
#W5A1
def Max_Of_Two(a, b) -> int:
    if a > b:
        return a
    return b
```

```python
#W5A2
def swap(a, b) -> tuple[int, int]:
    a = a^b
    b = a^b
    a = a^b
    return (a, b)
```

```python
#W5A3
def prime(a):
    if a < 2:
        return False
    ans = True
    for i in range(2, a):
        if a%i == 0:
            ans = False
            break
    return ans
```

```python
#W5A4
def perfect(n):
    A = []
    for i in range(1, n):
        if n%i == 0:
            A.append(i)
    ans = 0
    for i in range(len(A)):
        ans += A[i]
    return ans == n
```

```python
#W5A5
def find(A, k):
    ans = -1
    for i in range(len(A)):
        if A[i] == k:
            ans = i
            break
    return ans
```

```python
#W5A6
def frac(n):
    ans = 1
    for i in range(1, n+1):
        ans *= i
    return ans
```

```python
#W5A9
n = int(input())
ans = 0
while n > 0:
    ans += n%10
    n //= 10
print(ans)
```

```python
#W5A10
X, Y = input().split()
def mapping(X, Y):
    A = list(X)
    B = list(Y)

    n = len(A)
    M = [True]*n

    for i in range (n):
        temp = A[i]
        for j in range (i, n):
            if M[j] and A[j] == temp:
                A[j] = B[i]
                M[j] = False
    return A == B
if mapping(X, Y) and mapping(Y, X):
    print("true")
else:
    print("false")
```
