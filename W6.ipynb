```python
#W6A10
A = list(map(int, input().split()))
n = int(input())
S = set()
for i in range(len(A)):
    for j in range(i+1, len(A)):
       if n == A[i]+A[j]:
            S.add((A[i], A[j]))
S = sorted(list(S))
print(S)
```

```python
#W6A12
A = list(map(int, input().split()))
D = {}
for i in range(len(A)):
    if A[i] not in D:
        D[A[i]] = 0
    D[A[i]] += 1
ans = []
max = -1000000
for i, j in D.items():
    if j > max:
        max = j
for i, j in D.items():
    if j == max:
        ans.append(i)
print(min(ans))
    
```

```python
#W6A18
m, n, k = tuple(map(int, input().split()))
k = k-1
A = []
for i in range(n):
    A.append(list(map(int, input().split())))
ans = 0
for i in range(m):
    for j in range(n):
        if i == k or j == k:
            ans += A[i][j]
print(ans)
```
