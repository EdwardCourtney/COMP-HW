```python
#W10A3

S = set()

def num(a):
    ans = 0
    while a>0:
        ans += (a%10)**2
        a //= 10
    
    if ans in S:
        return False
    
    S.add(ans)
    
    if ans == 1:
        print(ans)
        return True
    else:
        print(ans)
        return num(ans)

a = int(input())
print(num(a))

```
